```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load your data from a CSV file (replace 'your_data.csv' with your actual file path)
data = pd.read_csv(r"C:\Users\Admin\Downloads\USvideos.csv")
```

```python
# Distribution of variables (using histograms)
data.hist(figsize=(10, 10))
plt.xlabel('Variable')
plt.ylabel('Frequency')
plt.title('Distribution of variables')
plt.subplots_adjust(bottom=0.1)
plt.show()
```

```python
# Identify outliers (using boxplots)
sns.boxplot(data=data, orient="v")  # For horizontal boxplots use 'orient="h"'
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Outlier detection (boxplots)')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

```python
# Correlation between variables (using heatmap)
correlation = data.corr(numeric_only=True)
sns.heatmap(correlation, annot=True)
plt.title('Correlation matrix')
plt.show()
```

```python
# Analyze specific variable correlations (e.g., likes vs views)
numerics = data.select_dtypes(include=[np.number])
sns.jointplot(x='views', y='likes', data=numerics)  # Replace 'views' and 'likes' with your desired variables
plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Correlation between Views and Likes')
plt.show()
```
